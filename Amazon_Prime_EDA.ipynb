```python
titles = titles.dropna(subset=['imdb_score'])
```

```python
titles.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='type', data=titles)

plt.title("Movies vs TV Shows on Amazon Prime")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

Visualization 1: Distribution of Movies vs TV Shows

This bar chart shows the number of Movies and TV Shows available on Amazon Prime.

Insight:
From the graph, we can see that Movies are significantly more numerous than TV Shows in the dataset. This indicates that Amazon Prime has a larger collection of movies compared to television series. The platform appears to focus more on providing movie content to its audience.

```python
titles['release_year'].value_counts().sort_index().plot(figsize=(10,5))

plt.title("Number of Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

Visualization 2: Titles Released Per Year

This line graph shows how many titles were released each year on Amazon Prime.

Insight:
From the graph, we can observe that the number of titles released has increased significantly in recent years. This trend reflects the rapid growth of streaming platforms and the increasing demand for digital entertainment content worldwide.

```python
sns.histplot(titles['imdb_score'], bins=20)

plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Frequency")

plt.show()
```

Visualization 3: IMDb Rating Distribution

This histogram shows the distribution of IMDb ratings for titles available on Amazon Prime.

Insight:
From the graph, we can observe that most titles have IMDb ratings between 6 and 7. This suggests that the majority of the content on Amazon Prime is moderately well-rated by viewers.

```python
titles['genres'].value_counts().head(10).plot(kind='bar', figsize=(10,5))

plt.title("Top 10 Genres on Amazon Prime")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")

plt.show()
```

Visualization 4: Top Genres on Amazon Prime

This bar chart shows the most common genres among titles available on Amazon Prime.

Insight:
From the graph, we can see that certain genres appear more frequently than others. Genres such as drama, comedy, and action are among the most common. This indicates that these genres are popular among viewers and widely produced by content creators.

```python
top_titles = titles.sort_values(by='imdb_score', ascending=False).head(10)

sns.barplot(x='imdb_score', y='title', data=top_titles)

plt.title("Top Rated Titles on Amazon Prime")

plt.show()
```

Visualization 5: Top Rated Titles on Amazon Prime

This bar chart displays the titles with the highest IMDb ratings available on Amazon Prime.

Insight:
These titles represent some of the highest rated content on the platform. High ratings indicate strong audience appreciation and critical acclaim for these shows or movies.
